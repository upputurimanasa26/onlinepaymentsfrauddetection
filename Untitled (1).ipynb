```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def transformationPlot(feature):
    # Everything inside the function must be indented!
    plt.figure(figsize=(12, 5))
    
    # Plot 1: Distribution
    plt.subplot(1, 2, 1)
    sns.histplot(feature, kde=True) # sns.distplot is deprecated; use histplot
    plt.title("Distribution Plot")
    
    # Plot 2: Probability Plot
    plt.subplot(1, 2, 2)
    # Fix: use '=' instead of '-' and ensure plot=plt is correct
    stats.probplot(feature, plot=plt) 
    plt.title("Probability Plot")
    
    plt.show()
```

```python
from flask import Flask, render_template, request
import pickle
import numpy as np

app = Flask(__name__)

# 1. Load your saved model (Ensure payments.pkl is in the same folder as this notebook)
model = pickle.load(open('payments.pkl', 'rb'))

@app.route('/')
def home():
    return render_template('home.html')

@app.route('/predict')
def predict_page():
    return render_template('predict.html')

@app.route('/submit', methods=['POST'])
def submit():
    # 2. Get data from the predict.html form
    # Note: Ensure the names match your input fields exactly
    features = [float(x) for x in request.form.values()]
    final_features = [np.array(features)]
    
    # 3. Make prediction
    prediction = model.predict(final_features)
    
    output = "FRAUD" if prediction[0] == 1 else "LEGITIMATE"
    
    return render_template('submit.html', prediction_text=f'Transaction is {output}')

if __name__ == "__main__":
    # Use port 5000 (default)
    app.run(port=5000)
```
